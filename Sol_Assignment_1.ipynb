```python
import numpy as np
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

```python
K=float(input("Please enter the strike price: "))
P=float(input("Please enter the premium: "))
S=np.arange(K-100,K+101,0.5)
Buy_Call_Profit=np.maximum(0,S-K)-P
Sell_Call_Profit=-Buy_Call_Profit
Buy_Put_Profit=np.maximum(0,K-S)-P
Sell_Put_Profit=-Buy_Put_Profit

traces = [
    go.Scatter(x=S, y=Buy_Call_Profit, mode='lines', name='Buying Call', line=dict(color='blue')),

    go.Scatter(x=S, y=Sell_Call_Profit, mode='lines', name='Selling Call', line=dict(color='red')),

    go.Scatter(x=S, y=Buy_Put_Profit, mode='lines', name='Buying Put', line=dict(color='green')),

    go.Scatter(x=S, y=Sell_Put_Profit, mode='lines', name='Selling Put', line=dict(color='orange')),

    #to highlight breakeven line
    go.Scatter(x=S, y=[0]*len(S), mode='lines', name='Break-even', line=dict(color='black', dash='dot'))

]

layout = go.Layout(
    title='Options Profit vs SpotPrice',
    xaxis_title='Spot Price',
    yaxis_title='Profit / Loss',
    hovermode='x unified',
    template='plotly_white'
)


fig = go.Figure(data=traces, layout=layout)
fig.show()


#We have the following inference: The maximum loss is limited in case of buy call and buy put , whereas the maximum profit is bound in the case of sell call and sell put, however this doesn't mean that only buying is safe.
```

```python
plt.figure(figsize=(20, 8))

plt.plot(S, Buy_Call_Profit, label='Buying Call', color='blue')
plt.plot(S, Sell_Call_Profit, label='Selling Call', color='red')
plt.plot(S, Buy_Put_Profit, label='Buying Put', color='green')
plt.plot(S, Sell_Put_Profit, label='Selling Put', color='orange')

plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

plt.axvline(K, color='gray', linewidth=0.5, linestyle='--')

plt.title('Profit/Loss vs Spot Price for Options')

plt.xlabel('Spot Price')

plt.ylabel('Profit / Loss')

plt.legend()
plt.grid(True)
plt.show()
```
